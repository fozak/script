```python
import zipfile

with zipfile.ZipFile('website_download.zip', 'r') as zip_ref:
    # List all files in the zip
    zip_info_list = zip_ref.infolist()
    for zip_info in zip_info_list:
        print(f"File Name: {zip_info.filename}")
        print(f"Uncompressed Size: {zip_info.file_size} bytes")
        print(f"Compressed Size: {zip_info.compress_size} bytes")
        print(f"Compression Method: {zip_info.compress_type}")
        print(f"Last Modified: {zip_info.date_time}")
        print("-" * 40)
```

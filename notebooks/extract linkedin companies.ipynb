```python
from bs4 import BeautifulSoup

# Path to your HTML file
file_path = 'C:/Users/i7714/Downloads/Search_Sales_Navigator.html'

# Read the HTML file
with open(file_path, 'r', encoding='utf-8') as file:
    html_content = file.read()

# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find all <a> tags that contain "company" in their href attributes
company_links = soup.find_all('a', href=lambda href: href and 'company' in href)

# List to hold the results
results = []

# Iterate through each company link and find associated person link and title
for company_link in company_links:
    # Find parent div of the company link
    parent_div = company_link.find_parent('div', class_='artdeco-entity-lockup__content ember-view')
    
    if parent_div:
        # Check for a sibling div that contains a person's link
        person_div = parent_div.find('div', class_='artdeco-entity-lockup__title ember-view')
        
        if person_div:
            person_link = person_div.find('a', href=lambda href: href and 'lead' in href)
            if person_link:
                # Get the person's name
                person_name = person_link.get_text(strip=True)

                # Search for any title element within the same parent div
                title_span = parent_div.find('span', {'data-anonymize': 'title'})
                title = title_span.get_text(strip=True) if title_span else "No title found"
                
                # Collect the details
                company_name = company_link.get_text(strip=True)
                results.append((person_name, person_link['href'], company_name, company_link['href'], title))

# Print the results
if results:
    for person_name, person_href, company_name, company_href, title in results:
        print(f'Person Name: {person_name}')
        print(f'Person Href: {person_href}')
        print(f'Company Name: {company_name}')
        print(f'Company Href: {company_href}')
        print(f'Title: {title}')
        print('-' * 40)
else:
    print("No links containing 'person' associated with 'company' found.")
```

```python
import requests
from bs4 import BeautifulSoup

# URL to fetch the HTML content from
url = 'https://fozak.github.io/script/notebooks/Search_Sales_Navigator_3.html'

# Fetch the HTML content from the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    html_content = response.text  # Get the HTML content
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
    html_content = None

# Proceed only if we successfully fetched the HTML
if html_content:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

# Find all <a> tags that contain "company" in their href attributes
company_links = soup.find_all('a', href=lambda href: href and 'company' in href)

# List to hold the results
results = []

# Iterate through each company link and find associated person link
for company_link in company_links:
    # Find parent div of the company link
    parent_div = company_link.find_parent('div', class_='artdeco-entity-lockup__content ember-view')
    
    if parent_div:
        # Check for a sibling div that contains a person's link
        person_div = parent_div.find('div', class_='artdeco-entity-lockup__title ember-view')
        
        if person_div:
            person_link = person_div.find('a', href=lambda href: href and 'lead' in href)
            if person_link:
                # Collect the details
                person_name = person_link.get_text(strip=True)
                company_name = company_link.get_text(strip=True)
                results.append((person_name, person_link['href'], company_name, company_link['href']))

# Print the results
if results:
    for person_name, person_href, company_name, company_href in results:
        print(f'Person Name: {person_name}')
        print(f'Person Href: {person_href}')
        print(f'Company Name: {company_name}')
        print(f'Company Href: {company_href}')
        print('-' * 40)
else:
    print("No links containing 'person' associated with 'company' found.")
```

```python
import urllib.parse
import json

# The full URL from LinkedIn Sales Navigator
url = savepage_url
# Parse the URL
parsed_url = urllib.parse.urlparse(url)
query_params = urllib.parse.parse_qs(parsed_url.query)

# Extract the query parameter and decode it
query_string = query_params['query'][0]

# Manually parse the query string (since it's a complex nested structure)
def parse_linkedin_query(query_str):
    # Remove surrounding parentheses
    query_str = query_str.strip('()')
    
    # Create a dictionary to store parsed results
    parsed_query = {}
    
    # Parse spellCorrectionEnabled
    parsed_query['spellCorrectionEnabled'] = True
    
    # Parse recentSearchParam
    if 'recentSearchParam:' in query_str:
        recent_search_match = query_str.split('recentSearchParam:')[1].split(',filters:')[0].strip('()')
        parsed_query['recentSearchParam'] = {
            'id': int(recent_search_match.split('id:')[1].split(',')[0]),
            'doLogHistory': True
        }
    
    # Parse filters
    filters_str = query_str.split('filters:List(')[1].split('),keywords:')[0]
    parsed_query['filters'] = []
    
    # Helper function to parse individual filter values
    def parse_filter_values(values_str):
        values = []
        # Remove outer List() and split individual values
        values_list = values_str.strip('List()').split('),(')
        for val in values_list:
            # Parse each value
            val_dict = {}
            val_parts = val.split(',')
            for part in val_parts:
                if ':' in part:
                    key, value = part.split(':')
                    # Try to convert to int if possible, else keep as string
                    try:
                        val_dict[key] = int(value)
                    except ValueError:
                        val_dict[key] = value.strip('"')
            values.append(val_dict)
        return values
    
    # Split different filter types
    filter_types = filters_str.split('),(')
    for filter_type in filter_types:
        if 'type:' in filter_type and 'values:' in filter_type:
            # Extract filter details
            type_match = filter_type.split('type:')[1].split(',values:')[0]
            values_match = filter_type.split('values:')[1]
            
            parsed_filter = {
                'type': type_match,
                'values': parse_filter_values(values_match)
            }
            parsed_query['filters'].append(parsed_filter)
    
    # Parse keywords
    parsed_query['keywords'] = query_params.get('query', [''])[0].split('keywords:')[-1]
    
    return parsed_query

# Parse the query and convert to JSON
parsed_result = parse_linkedin_query(query_string)

# Print the JSON
print(json.dumps(parsed_result, indent=2))
```

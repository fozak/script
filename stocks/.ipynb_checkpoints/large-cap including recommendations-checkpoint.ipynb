```python
import yfinance as yf

# List of tickers
tickers = [
    "GRAB", "UBER", "CFLT", "MDB", "PCOR", "DASH", "OKTA", "AFRM",
    "CART", "PCTY", "PAYC", "DT", "PSN", "DAY", "TWLO", "ESTC",
    "WDAY", "SPOT", "GEN", "NTNX", "CPAY"
]

# Function to get last four quarters of revenue
def get_last_four_quarters_revenue(tickers):
    revenue_data = {}
    
    for ticker in tickers:
        stock = yf.Ticker(ticker)
        # Get the quarterly financials
        quarterly_financials = stock.quarterly_financials
        
        # Extract the revenue for the last four quarters
        if 'Total Revenue' in quarterly_financials.index:
            revenue_data[ticker] = quarterly_financials.loc['Total Revenue'].head(4)
        else:
            revenue_data[ticker] = "Revenue data not available"
    
    return revenue_data

# Get the revenue data
revenue_data = get_last_four_quarters_revenue(tickers)

# Print the results
for ticker, revenue in revenue_data.items():
    print(f"{ticker}: {revenue}")
```

```python
# Import the yfinance library
import yfinance as yf
import pandas as pd

# List of tickers
tickers = [
    "DOCN",
    "APPN",
    "ODD",
    "SRAD",
    "PAYO",
    "RELY",
    "FIVN",
    "EVCM",
    "ZI",
    "UPWK",
    "RPD",
    "CXM",
    "FOUR",
    "CALX",
    "WEX",
    "LSPD",
    "GBTG",
    "RNG",
    "OTEX",
    "LYFT"
]

# Function to get analyst recommendations
def get_analyst_recommendations(tickers):
    recommendations = {}
    for ticker in tickers:
        try:
            stock = yf.Ticker(ticker)
            # Get recommendations
            recs = stock.recommendations
            # Print the recommendations DataFrame for debugging
            print(f"Recommendations for {ticker}:\n", recs.head(), "\n")
            
            if recs is not None and not recs.empty:
                # Count recommendations
                if 'Recommendation' in recs.columns:
                    rec_counts = recs['Recommendation'].value_counts().to_dict()
                    recommendations[ticker] = rec_counts
                else:
                    recommendations[ticker] = "No 'Recommendation' column found"
            else:
                recommendations[ticker] = "No recommendations available"
        except Exception as e:
            recommendations[ticker] = str(e)
    return recommendations

# Get analyst recommendations
analyst_recs = get_analyst_recommendations(tickers)

# Display the recommendations
for ticker, rec in analyst_recs.items():
    print(f"{ticker}: {rec}")
```

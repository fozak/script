```python
import sqlite3
import re

def clean_mysql_sql(sql):
    # Remove MySQL-specific comments
    sql = re.sub(r'/\*![0-9]+ .*?\*/;', '', sql, flags=re.DOTALL)
    # Remove ENGINE and CHARSET statements
    sql = re.sub(r'ENGINE=\w+ DEFAULT CHARSET=\w+ COLLATE=\w+ ROW_FORMAT=\w+;', ';', sql)
    # Replace MySQL-style backticks with double quotes (SQLite compatible)
    sql = sql.replace('`', '"')
    return sql

# Step 1: Load and clean the original MySQL SQL dump
with open('erp.sql', 'r', encoding='utf-8') as f:
    raw_sql = f.read()

clean_sql = clean_mysql_sql(raw_sql)

# Step 2: Create a new SQLite database with a different name
output_db_name = 'erp_converted.db'
conn = sqlite3.connect(output_db_name)
cursor = conn.cursor()

# Step 3: Split and execute SQL statements
for statement in clean_sql.split(';'):
    stmt = statement.strip()
    if stmt:
        try:
            cursor.execute(stmt)
        except Exception as e:
            print(f"⚠️ Skipped statement due to error:\n{e}\n{stmt[:100]}...\n")

# Step 4: Commit and close
conn.commit()
cursor.close()
conn.close()

print(f"✅ Export complete. SQLite database saved as {output_db_name}.")
```

```python
import sqlite3

# Define the database and target project
db_path = 'erp_converted.db'
project_name = 'Inspection'  # Change to your actual project name

# Connect to the SQLite database
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Query all tasks for the given project
try:
    cursor.execute("""
        SELECT name, subject, status
        FROM "tabTask"
        WHERE project = ?
    """, (project_name,))
    
    tasks = cursor.fetchall()
    
    if tasks:
        print(f"✅ Tasks for project: {project_name}")
        for name, subject, status in tasks:
            print(f"- {name} | {subject} | Status: {status}")
    else:
        print(f"⚠️ No tasks found for project: {project_name}")
except sqlite3.Error as e:
    print(f"❌ Database error: {e}")

# Close the connection
cursor.close()
conn.close()
```
